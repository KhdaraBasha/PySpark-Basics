## SELECT METHOD FROM PySpark

```python
# Path to data file
path = '/Volumes/workspace/practice_db/sample_data/emp_1.csv'

# Set options for reading CSV
header = True        # Use first row as header
infer_schema = True   # Automatically infer column data types

# Read data into a DataFrame 
emp_df = spark.read.csv(path, header = header, inferSchema = infer_schema)

# Display the DataFrame
emp_df.show(2)

# Display the schema of the DataFrame
emp_df.printSchema()
```

```python
emp_df.select(['name']).dropDuplicates().show()
```

```python
emp_df.select(['emp_id', 'name']).dropDuplicates().show()
```

```python
emp_df.select(emp_df.emp_id, emp_df.name, emp_df.salary).dropDuplicates().show()
```

```python
emp_df.select(emp_df.columns[0:5]).dropDuplicates().show()
```
