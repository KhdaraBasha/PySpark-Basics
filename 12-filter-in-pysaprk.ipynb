## FILTER METHOD IN PYSPARK

```python
# Define the path to the CSV file
path = '/Volumes/workspace/practice_db/sample_data/emp_1.csv'

# Read the CSV file into a Spark DataFrame with header and schema inference
df = spark.read.csv(path, header=True, inferSchema=True)

# Display the DataFrame in Databricks
display(df)

# Print the schema of the DataFrame
df.printSchema()
```

```python
df.filter(df.address == 'india').show()
```

```python
df.filter(df.salary.isNull()).show()
```

```python
df.filter((df.address == 'india') & (df.salary >= 5000)).dropDuplicates().show()
```

```python
df.filter((df.address != 'india') & (df.salary >= 5000)).dropDuplicates().show()
```

```python
df.filter((df.address == 'india') | (df.salary >= 5000)).dropDuplicates().show()
```

```python
from pyspark.sql.functions import col, lit, lower
df.filter(lower(col("name")).startswith('m')).dropDuplicates().show()
```

```python
df.filter(df.name.like('%ul%')).dropDuplicates().show()
```
