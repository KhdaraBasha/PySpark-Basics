## Read Data From CSV Files with Different Approaches

```python
## Path to the employee data CSV file
file_path = '/Volumes/workspace/practice_db/sample_data/emp_1.csv'

## Set the options for reading the CSV file
header = False
infer_schema = False
```

```python
# Read the CSV file into a Spark DataFrame
df = spark.read.csv(file_path, header=header, inferSchema=infer_schema)
# Display the first two rows of the DataFrame
df.show(2)
# Display the schema of the DataFrame
df.printSchema()
```

```python
## Set the options for reading the CSV file
header = True
infer_schema = True

# Read the CSV file into a Spark DataFrame
df1 = spark.read.csv(file_path, header=header, inferSchema=infer_schema)
# Display the first two rows of the DataFrame
df1.show(2)
# Display the schema of the DataFrame
df1.printSchema()
```

```python
# Read the CSV file into a Spark DataFrame using the format() method
df3 = spark.read.format('csv').option('header', True).load(file_path)
# Display the first two rows of the DataFrame
df3.show(2)
# Display the schema of the DataFrame
df3.printSchema()
```

```python
# Read the CSV file into a Spark DataFrame using the format() method and chain the options method to set the options for inferring the schema and specifying the header row 
df4 = spark.read.format('csv')\
    .option('header',True)\
    .option('inferSchema', True)\
    .load(file_path)
# Display the first two rows of the DataFrame
df4.show(2)
# Display the schema of the DataFrame
df4.printSchema()
```

```python
# Read the CSV file into a Spark DataFrame using the format() method and chain the options method to set the options for inferring the schema and specifying the header row
df5 = spark.read.format('csv').options(header=True, inferSchema=True).load(file_path)
# Display the first two rows of the DataFrame 
df5.show(2)
# Display the schema of the DataFrame
df5.printSchema()
```
