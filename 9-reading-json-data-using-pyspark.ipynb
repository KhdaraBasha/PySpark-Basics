## Reading JSON data using PySpark

```python
# File path to read the JSON
file_path = '/Volumes/workspace/practice_db/sample_data/json_example1.json'
```

```python
# Read the JSON file into a DataFrame using the SparkSession method `read.json`
df = spark.read.json(file_path)
# Display the DataFrame
df.show()
# Display the schema
df.printSchema()
```
